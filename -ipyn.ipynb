```python
# 데이터 추출
categories = data.iloc[0, 1:].tolist()  # 첫 번째 행의 데이터를 카테고리로 추출합니다.
avg_temp = data.iloc[1, 1:].astype(float)  # 두 번째 행의 데이터를 평균 기온으로 추출합니다. 데이터는 float 타입으로 변환합니다.
max_temp = data.iloc[2, 1:].astype(float)  # 세 번째 행의 데이터를 최고 기온으로 추출합니다. 데이터는 float 타입으로 변환합니다.
min_temp = data.iloc[3, 1:].astype(float)  # 네 번째 행의 데이터를 최저 기온으로 추출합니다. 데이터는 float 타입으로 변환합니다.
rainfall = data.iloc[4, 1:].astype(float)  # 다섯 번째 행의 데이터를 강수량으로 추출합니다. 데이터는 float 타입으로 변환합니다.
sunshine = data.iloc[5, 1:].astype(float)  # 여섯 번째 행의 데이터를 일조량으로 추출합니다. 데이터는 float 타입으로 변환합니다.

# 그래프 그리기
plt.figure(figsize=(12, 8))  # 그래프의 크기를 설정합니다.
plt.rc('font', family='NanumGothic')  # 그래프의 폰트를 'NanumGothic'으로 설정합니다.

# 평균 기온 그래프
plt.subplot(2, 2, 1)  # 2x2 그리드의 첫 번째 위치에 그래프를 그립니다.
plt.bar(categories, avg_temp)  # 카테고리별로 평균 기온을 막대 그래프로 그립니다.
plt.title('평균 기온')  # 그래프의 제목을 설정합니다.
plt.ylabel('기온 (℃)')  # y축의 레이블을 설정합니다.
plt.xticks(rotation=45, ha='right')  # x축 눈금 레이블을 45도로 회전하고 오른쪽 정렬합니다.

# 최고 기온 그래프
plt.subplot(2, 2, 2)  # 2x2 그리드의 두 번째 위치에 그래프를 그립니다.
plt.bar(categories, max_temp)  # 카테고리별로 최고 기온을 막대 그래프로 그립니다.
plt.title('최고 기온')  # 그래프의 제목을 설정합니다.
plt.ylabel('기온 (℃)')  # y축의 레이블을 설정합니다.
plt.xticks(rotation=45, ha='right')  # x축 눈금 레이블을 45도로 회전

# 최저 기온 그래프
plt.subplot(2, 2, 3)  # 2x2 그리드의 세 번째 위치에 그래프를 그립니다.
plt.bar(categories, min_temp)  # 카테고리별로 최저 기온을 막대 그래프로 그립니다.
plt.title('최저 기온')  # 그래프의 제목을 설정합니다.
plt.ylabel('기온 (℃)')  # y축의 레이블을 설정합니다.
plt.xticks(rotation=45, ha='right')  # x축 눈금 레이블을 45도로 회전하고 오른쪽 정렬합니다.

# 강수량 그래프
plt.subplot(2, 2, 4)  # 2x2 그리드의 네 번째 위치에 그래프를 그립니다.
plt.bar(categories, rainfall)  # 카테고리별로 강수량을 막대 그래프로 그립니다.
plt.title('강수량')  # 그래프의 제목을 설정합니다.
plt.ylabel('강수량 (mm)')  # y축의 레이블을 설정합니다.
plt.xticks(rotation=45, ha='right')  # x축 눈금 레이블을 45도로 회전하고 오른쪽 정렬합니다.

plt.tight_layout()  # 그래프들을 자동으로 조정하여 겹치지 않도록 합니다.
plt.show()  # 그래프를 표시합니다.
```
